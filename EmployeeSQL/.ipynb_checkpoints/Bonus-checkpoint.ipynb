```python
import pandas as pd
import os
import psycopg2
from sqlalchemy import create_engine
import sqlalchemy as db
import matplotlib.pyplot as plt
engine = create_engine('postgresql+psycopg2://postgres:admin@localhost/Homework')
connection = engine.connect()
```

```python
# Import the Employees table
employees = pd.read_sql('select * from employees', connection)
employees.head()
```

```python
# Import the Salaries table
salaries = pd.read_sql('select * from salaries', connection)
salaries.head()
```

```python
# Import the Titles table
titles = pd.read_sql('select * from titles', connection)
titles.head()
titles_df = titles.rename(columns = {"title_id": "emp_title_id"})
titles_df
```

```python
merge_df=employees.merge(titles_df, on="emp_title_id")
merge_df.head()
```

```python
Final_df=merge_df.merge(salaries, on="emp_no")
Final_df.head()
```

```python
employee_salaries_df = Final_df[['emp_no', 'title', 'salary']]
employee_salaries_df.head()
```

```python
employee_salaries_df.hist(column='salary')
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Salary Range')
plt.ylabel('Frequency')
plt.title('Most Common Salary Ranges')
plt.savefig("Most Common Salary Ranges")
plt.show()
```

```python
employees_grouped_by_title = employee_salaries_df.groupby(['title'])['salary'].mean()
Clean_df = pd.DataFrame(employees_grouped_by_title)
Clean_df = Clean_df.sort_values("salary", ascending = False)
Clean_df
```

```python
Clean_df.plot.bar()
plt.title("Average Salary by Title", fontsize = 14)
plt.xlabel("Title")
plt.ylabel("Average Salary")
plt.savefig("Average Salary by Title")
plt.show()
```
